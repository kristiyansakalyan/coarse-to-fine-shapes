```python
from visuals.visuals_util import visualize_pointcloud_eval, visualize_pointcloud_eval_three, visualize_pointcloud
import torch
from pathlib import Path
```

```python
def load_pts_to_tensor(filename):
    with open(filename, 'r') as file:
        points = []
        for line in file:
            x, y, z = map(float, line.strip().split())
            points.append([x, y, z])
        return torch.tensor(points, dtype=torch.float32)
```

```python
folder_path = Path("/Users/stephanschmiedmayer/Downloads/visuals/point-transformer/2024-01-30_23-46-44")

for i in range(10):
    index = f"{i:02d}"
    input = load_pts_to_tensor(folder_path / f"{index}-input.pts")
    refined_output = load_pts_to_tensor(folder_path / f"{index}-refined_output.pts")
    fig = visualize_pointcloud_eval(input, refined_output, index)
    fig.savefig(f"visuals/point-transformer/{index}.png", dpi=400)    
```
